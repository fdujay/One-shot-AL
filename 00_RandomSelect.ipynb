```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

import os
import glob
import random
import shutil
```

```python
# pics_dir = '../2048/Pics2048/'
# mask_dir = '../2048/Mask2048/'
pics_dir = '../2048/Train_Pics/'
mask_dir = '../2048/Train_Mask/'
query_pics_dir = '../2048/1000/VAE/Random/Pics/'
query_mask_dir = '../2048/1000/VAE/Random/Mask/' 
number = 1000
# ratio = 0.25
```

```python
def makedir(new_dir):
    if not os.path.exists(new_dir):
        os.makedirs(new_dir)
```

```python
# coding:utf-8
import os
'''
    获得图片名称
'''

def gen_list(img_dir):
    names = []
    img_list = os.listdir(img_dir)                    # 获取类别文件夹下所有png图片的路径
    for i in range(len(img_list)):
        if not img_list[i].endswith('jpg'):         # 若不是jpg文件，跳过
            continue
        name = img_list[i].split('.')[0]
        
        names.append(name)
        
    return names


ImgList = gen_list(pics_dir)
ImgList
```

```python
def random_query(full_list, number, shuffle=False):
    """
    数据集拆分: 将列表full_list按比例ratio（随机）划分为2个子列表sublist_1与sublist_2
    :param full_list: 数据列表
    :param ratio:     子列表1
    :param shuffle:   子列表2
    :return:
    """
    n_total = len(full_list)
#     offset = int(n_total * ratio)
    offset = number
    if n_total == 0 or offset < 1:
        return [], full_list
    if shuffle:
        random.shuffle(full_list)
    sublist_1 = full_list[:offset]
    sublist_2 = full_list[offset:]
    return sublist_1, sublist_2
 
 
if __name__ == "__main__":
    labeled_list, unlabeled_list = random_query(ImgList, number=number, shuffle=True)
    print(len(labeled_list))
    print(len(unlabeled_list))
```

```python
#创建图像文件夹
def gen_pic_file(in_dir, out_dir, names):
    makedir(out_dir)
    i=0
    for name in names:
        img_dir = os.path.join(in_dir, name + '.jpg')
    #     out_path = os.path.join(out_dir, os.path.split(imgs_list[i])[-1])
        out_path = out_dir
    #     print(sDir)
        shutil.copy(img_dir, out_path)
        i += 1
    print(i)
#     print('Class:{}, train:{}, valid:{}, test:{}'.format(sDir, train_point, valid_point-train_point, imgs_num-valid_point))
```

```python
gen_pic_file(pics_dir, query_pics_dir, labeled_list)
```

```python
#创建模板文件夹
def gen_mask_file(in_dir, out_dir, names):
    makedir(out_dir)
    i=0
    for name in names:
        img_dir = os.path.join(in_dir, name + '_segmentation.png')
    #     out_path = os.path.join(out_dir, os.path.split(imgs_list[i])[-1])
        out_path = out_dir
    #     print(sDir)
        shutil.copy(img_dir, out_path)
        i += 1
    print(i)
```

```python
gen_mask_file(mask_dir, query_mask_dir, labeled_list)
```
