```python
'''
ft:挑选后的特征
t:特征对应金标准
name:样本名称
c: 特征对应簇
# '''
# res_flag = False
# if res_flag:
#     df1 = pd.read_csv('../FeatureBank/CIFAR10_1.0EXP_VAE.csv',header = None)
#     print(df1.shape)
#     ft = df1.values #特征样本/256维
#     ft = scale(ft) #进行归一化

#     df2 = pd.read_csv('../cifar-10-1.0-exp/train_cluster_res.csv',header = 0)
#     # df2
#     name = df2.iloc[:,0].values
#     t = df2.iloc[:,1].values
#     c = df2.iloc[:,2].values
#     name = df2.iloc[:,0].values
#     print('Original dataset:', Counter(t))
#     print('Clustering:', Counter(c))
    
# else:
df1 = pd.read_csv('../FeatureBank/Caltech256_VAE.csv',header = None)
# print(df1.shape)
ft = df1.values #特征样本/256维
# ft = scale(ft) #进行归一化

df2 = pd.read_csv('../txt/train.txt',sep = ' ',header = None)
name = df2.iloc[:,0].values #图片位置名称
t = df2.iloc[:,1].values #金标准0/1
print('Original dataset:', Counter(t))

df3 = pd.read_csv('../kmeans/Caltech_256_kmeans_pca.csv',header = 0)
c = df3.iloc[:,1].values
print('Clustering:', Counter(c))
```

```python
n_clusters = 256
ratio=0.1
total_num = len(t)
sel_num= math.ceil(ratio*total_num/n_clusters)
```

```python
'''
隔离森林local IFL
'''
import math
from sklearn.ensemble import IsolationForest

index=[] #index[i]表示第i个聚类对应的index列表
ft_part=[] #ft_part[i]表示第i个聚类对应的特征向量列表
new_index=[] #新列表

for i in range(n_clusters):
    index.append(np.where(c == i)[0])
    ft_part.append(ft[index[i]])

for i in range(n_clusters):
    #注意：比例
    clf = IsolationForest(contamination=ratio,n_jobs=20).fit(ft_part[i])
    
    novel_pred = clf.predict(ft_part[i])
    li = np.where(novel_pred==-1)
    
#     novel_scores = clf.score_samples(ft_part[i])
#     li = np.argsort(novel_scores)[:sel_num]
    new_index.extend(index[i][li])


ft_sel = ft[new_index]
t_sel = t[new_index]
c_sel = c[new_index]
name_sel = name[new_index]

from collections import Counter 
print('Resampled dataset:', Counter(t_sel))
print(len(t_sel))

a=dict(Counter(t_sel))
np.mean(list(a.values())), np.var(list(a.values()))
```

```python
'''
注意保存csv的后缀
'''
print(__doc__)
dataframe1 = pd.DataFrame({'image_name':name_sel,'target':t_sel})
dataframe1.to_csv("../txt/Caltech256_IFL10_pca.txt",index=False,sep=' ',header = None)
print(Counter(t_sel))
```
